```python
#if you need to see about the socialmedia of particular country normalize give the data in percentage 
#country_grp['SocialMedia'].value_counts(normalize=True).loc['Russian Federation']
```

```python
df.groupby('Country')['ConvertedCompYearly'].median().loc['Germany']
#this is used to see the median of specific country 
#.loc is used to see about one paricular city
```

```python
df.groupby('Country')['ConvertedCompYearly'].agg(['median','mean'])
#.agg used to see mean and median at the same time
```

```python
#in this we chaeck the how many people from india know pythoon and finally we get to know using the sum function 
filt=df['Country']=='India'
df.loc[filt]['LanguageHaveWorkedWith'].str.contains('Python').sum()
```

```python
#what percentage of people from each country know python we cannot apply string operations to groupby functions
(df['LanguageHaveWorkedWith'].str.contains('Python')
                               .groupby(df['Country'])
                               .value_counts(normalize=True)
)
```
